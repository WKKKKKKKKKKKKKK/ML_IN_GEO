```python
import numpy as np
import torch
```

```python
def l(x):
    sum=0
    for i in range(len(x)):
        sum+=(1/len(x))*((x[i]+8)**4+6)
    return sum
```

```python
x=np.array([1,3,5,7,9])

x_tensor = torch.tensor(x, dtype=torch.float32, requires_grad=True)
y = l(x_tensor)
y.backward()
print(x_tensor.grad)
```
